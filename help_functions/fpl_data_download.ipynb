```python
'''
events = gameweeks, in FPL
game_settings = fpl game settings
phases = months, and which game days they cover
teams = as titled
total_players = # of total fpl managers this season
elements = players in the premier league
element_stats = metadata of certain player statistics i.e. minutes = Minutes played
element_types = information about different player positions (midfield, goalie)
'''
```

```python
players = pd.DataFrame(json['elements'])
player_types = pd.DataFrame(json['element_types'])
teams = pd.DataFrame(json['teams'])
events = pd.DataFrame(json['events'])
#element_stats = pd.DataFrame(json['element_stats'])
#phases = pd.DataFrame(json['phases'])

fixtures = pd.DataFrame(json2)
```

```python
teams.head().T
```

```python
events.columns
```

```python
events.head().T
```

```python
events['deadline_time'] = pd.to_datetime(events['deadline_time']).dt.date
```

```python
#store the last gameweek as a string (i.e. Gameweek_3); for labelling the data exported from this file
last_gameweek = events.loc[events['deadline_time'] < date.today()]['name'].iloc[-1].replace(' ', '_')
last_gameweek_number = events.loc[events['deadline_time'] < date.today()]['id'].iloc[-1]
```

```python
fixtures.drop(['id', 'code', 'finished', 'finished_provisional', 'minutes', 'provisional_start_time', 'started', 'stats', 'pulse_id'], axis=1, inplace=True)
fixtures.rename(columns={'event': 'Gameweek'}, inplace=True)
fixtures.head()
```

```python
fixtures['kickoff_date'] = fixtures['kickoff_time'].apply(lambda x: x.split('T')[0])
fixtures['kickoff_time'] = fixtures['kickoff_time'].apply(lambda x: x.split('T')[1].replace('Z', ''))
fixtures = fixtures[['Gameweek', 'kickoff_date', 'kickoff_time', 'team_a', 'team_a_score', 'team_h', 'team_h_score', 'team_h_difficulty', 'team_a_difficulty']]
```

```python
fixtures.head()
```

```python
fixtures['team_a'] = fixtures['team_a'].map(teams.set_index('id').name)
fixtures['team_h'] = fixtures['team_h'].map(teams.set_index('id').name)
```

```python
fixtures.head()
```

```python
url_myteam = 'https://fantasy.premierleague.com/api/entry/155692/event/{}/picks/'.format(last_gameweek_number)
r_myteam = requests.get(url_myteam)
json_myteam = r_myteam.json()
```

```python
my_team = pd.DataFrame(json_myteam['picks'])
my_team['element'] = my_team['element'].map(players.set_index('id').web_name)
my_team.drop('position', axis=1, inplace=True)
my_team.rename(columns={'element': 'Players'}, inplace=True)
my_team
```

```python
players.columns
```

```python
player_types.head()
```

```python
players['position'] = players.element_type.map(player_types.set_index('id').singular_name)
players.head()
```

```python
players['team'] = players.team.map(teams.set_index('id').name)
players.head()
```

```python
players = players.drop(columns = ['id', 'first_name', 'second_name', 'chance_of_playing_next_round', 'chance_of_playing_this_round', 'code', 'cost_change_event', 'cost_change_event_fall', 
                                  'cost_change_start', 'cost_change_start_fall', 'dreamteam_count', 'element_type', 'ep_next', 'ep_this',
                                  'form', 'in_dreamteam', 'news', 'news_added', 'photo', 'special', 'squad_number', 'team_code', 'transfers_in',
                                  'transfers_in_event', 'transfers_out', 'transfers_out_event', 'value_form', 'value_season', 'influence_rank',
                                  'influence_rank_type', 'creativity_rank', 'creativity_rank_type', 'threat_rank', 'threat_rank_type', 'ict_index_rank',
                                  'ict_index_rank_type', 'corners_and_indirect_freekicks_text', 'direct_freekicks_text', 'penalties_text', 'now_cost_rank', 
                                  'now_cost_rank_type', 'form_rank', 'form_rank_type', 'points_per_game_rank', 'points_per_game_rank_type', 'selected_rank', 
                                  'selected_rank_type'])
```
